```python
import pandas as pd
import numpy as np
from collections import Counter


from matplotlib import pyplot as plt
```

```python
lang = pd.read_csv("bar_data.csv")
```

```python
lang.head()
```

```python
lang.shape
```

```python
lang.LanguagesWorkedWith.nunique()
```

```python
lang["langs"] = lang["LanguagesWorkedWith"].apply(lambda x: x.split(";"))
```

```python
lang.head()
```

```python
lang.langs[0]
```

```python
total_list = []
```

```python
for i in range(len(lang.langs)):
    for k in lang.langs[i]:
        total_list.append(k)
```

```python
len(total_list)
```

```python
lang_most=Counter(total_list).most_common(15)
```

```python
lang_most
```

```python
x,y = zip(*lang_most)
x[:5]
```

```python
y[:5]
```

```python
x = list(x)
y = list(y)
```

```python
x.reverse()
y.reverse()
```

```python
fig,ax = plt.subplots(figsize=(15,8))
ax.barh(x,y)
plt.xticks(rotation=45)
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
rects1 = ax.bar(x, y);
ax.set_title("Most Popular Languages");
ax.set_ylabel('Number of People Who Use');
ax.set_xticks(x_pos);
plt.xticks(rotation=60)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
```
